```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

```python
# SKEWNESS , KURTOSIS #
```

```python
df = pd.read_csv('Q9_a.csv')
df
df.head()
```

```python
from scipy.stats import skew
```

```python
for col in df:
    print(col)
    print(skew(df[col]))
    
    plt.figure()
    sns.distplot(df[col])
    plt.show() 
    
    import warnings
    
    warnings.filterwarnings('ignore')
    %matplotlib inline
    
```

```python
from scipy.stats import kurtosis as kts
```

```python
for col in df:
    print(col)
    print(kts(df[col]))
    
    plt.figure()
    sns.distplot(df[col])
    plt.show() 
```
